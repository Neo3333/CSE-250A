```python
import numpy as np
import matplotlib.pyplot as plt
import math
%matplotlib inline
x = np.linspace(-100,100,1000)
y1 = np.log(np.cosh(x))
y2 = np.log(np.cosh(-2)) + ((np.exp(-2)-np.exp(2))/(np.exp(-2)+np.exp(2))) * (x+2) + 0.5 * (x+2) ** 2
y3 = np.log(np.cosh(3)) + ((np.exp(3)-np.exp(-3))/(np.exp(3)+np.exp(-3))) * (x-3) + 0.5 * (x-3) ** 2
```

#### Plots for $f(x)$ , $Q(x,-2)$ and $Q(x,3)$.

```python
fig=plt.figure(figsize=(30,10))
fig.add_subplot(1,3,1)
plt.plot(x,y1,'r')
plt.xlabel('x', fontdict={'family' : 'Times New Roman', 'size'   : 20})
plt.ylabel('f(x)', fontdict={'family' : 'Times New Roman', 'size'   : 20})
fig.add_subplot(1,3,2)
plt.plot(x,y2,'g')
plt.xlabel('x', fontdict={'family' : 'Times New Roman', 'size'   : 20})
plt.ylabel('Q(x,-2)', fontdict={'family' : 'Times New Roman', 'size'   : 20})
fig.add_subplot(1,3,3)
plt.plot(x,y3,'b')
plt.xlabel('x', fontdict={'family' : 'Times New Roman', 'size'   : 20})
plt.ylabel('Q(x,3)', fontdict={'family' : 'Times New Roman', 'size'   : 20})
plt.show()
```

```python
# update steps for x0 = -2
max_iteration = 50
x = -2
x1_list = [-2]
for i in range(max_iteration):
    x = x - ((np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)))
    x1_list.append(x)
x1_list = np.array(x1_list)
print(x1_list)
```

```python
# update steps for x0 = 3
max_iteration = 50
x = 3
x2_list = [3]
for i in range(max_iteration):
    x = x - ((np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)))
    x2_list.append(x)
x2_list = np.array(x2_list)
print(x2_list)
```

#### Plots for $x_n$ with respect to $n$ (number of iterations)

```python
x = np.linspace(0,x2_list.shape[0]-1,x2_list.shape[0])
fig=plt.figure(figsize=(30,10))
fig.add_subplot(1,2,1)
plt.plot(x,x1_list,'b')
plt.xlabel('number of iterations', fontdict={'family' : 'Times New Roman', 'size'   : 20})
plt.ylabel('x', fontdict={'family' : 'Times New Roman', 'size'   : 20})
fig.add_subplot(1,2,2)
plt.plot(x,x2_list,'g')
plt.xlabel('number of iterations', fontdict={'family' : 'Times New Roman', 'size'   : 20})
plt.ylabel('x', fontdict={'family' : 'Times New Roman', 'size'   : 20})
plt.show()
```

#### Using Newton Method, if x's initial value is too far away from the global minimum ($x_0 = -2$ or $x_0 = 3$), the update will not converge eventually.

```python
# x0 = -2
max_iteration = 20
x = -2
x3_list = [-2]
for i in range(max_iteration):
    x = x - 0.25*(np.exp(2*x)-np.exp(-2*x))
    x3_list.append(x)
x3_list = np.array(x3_list)
print(x3_list)
```

```python
# x0 = 3
max_iteration = 20
x = 1.09
x4_list = [1.09]
for i in range(max_iteration):
    x = x - 0.25*(np.exp(2*x)-np.exp(-2*x))
    x4_list.append(x)
x4_list = np.array(x4_list)
print(x4_list)
```

#### In order to converge we want the find the upper bound of $|x_0|$ so that after the first update, $|x_1| <= |x_0|$. 

```python
x = np.linspace(-5,5,10000)
y = x - 0.25 * (np.exp(2*x)-np.exp(-2*x))
z = np.abs(x)-np.abs(y)
index = np.where(z>0)
bound = np.abs(x[index][0])
print("upper bound for |x0| is %.4f"%bound)
```

#### Plots for $g(x) = \frac{1}{10}\sum_{k=1}^{10}log\ cosh(x+\frac{2}{\sqrt{k}})$.

```python
x = np.linspace(-100,100,1000)
y = np.zeros(1000)
for i in range(1,11):
    y = y + np.log(np.cosh(x+2/(i**0.5)))
y = y / 10
plt.plot(x,y,'b')
plt.xlabel('x', fontdict={'family' : 'Times New Roman', 'size'   : 20})
plt.ylabel('g(x)', fontdict={'family' : 'Times New Roman', 'size'   : 20})
plt.show()
```

#### Update rule $x_{n+1} = x_n - \frac{1}{10}\sum_{k=1}^{10}{sinh(x_n+\frac{2}{\sqrt{i}})*cosh(x_n+\frac{2}{\sqrt{i}})^{-1}}$

```python
max_iteration = 50
x = -10
x5_list = [-10]
for i in range(max_iteration):
    gradient = 0
    for j in range(1,11):
        gradient += np.sinh(x+2/j**0.5)/np.cosh(x+2/j**0.5)
    x = x - 1/10 * gradient
    x5_list.append(x)
x5_list = np.array(x5_list)
result = 0
for i in range(1,11):
    result = result + np.log(np.cosh(x5_list[-1]+2/(i**0.5)))
result = result / 10
print("the minimum of g(x) = %.4f when x = %.4f"%(result,x5_list[-1]))
```

#### Plots for $x_n$ with respect to $n$ (number of iterations)

```python
x = np.linspace(0,x5_list.shape[0]-1,x5_list.shape[0])
plt.plot(x,x5_list,'b')
plt.xlabel('number of iterations', fontdict={'family' : 'Times New Roman', 'size'   : 20})
plt.ylabel('Xn', fontdict={'family' : 'Times New Roman', 'size'   : 20})
plt.show()
```
